```python
def skew(text):
    skew_list = [0]
    for i in text:
        if i == 'C':
            skew_list.append(skew_list[-1]-1)
        elif i=='G':
            skew_list.append(skew_list[-1]+1)
        else:
            skew_list.append(skew_list[-1])
    return skew_list
    
```

```python
com_dict = {'A':'T','C':'G','G':'C','T':'A'}

def complement(pattern):
    array = ''
    for x in pattern:
        array += com_dict[x]
    return array[::-1]
```

```python
def minimum_skew_point(text):
    skew_list = skew(text)
    minimum_value = min(skew_list)
    minimum_indices = [i for i, a in enumerate(skew_list) if a == minimum_value]
    return minimum_indices
```

```python
def hammingDistance(p, q):
    count = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            count += 1
    return count
```

```python
def approximate_matching_pattern(pattern, text, k):
    positions = []
    for i in range(len(text)-len(pattern)+1):
        current_pattern = text[i:i+len(pattern)]
        if current_pattern == pattern or hammingDistance(current_pattern, pattern) <= k:
            positions.append(i)
    return ' '.join(str(x) for x in positions)
```

```python
def approximate_matching_pattern_count(pattern, text, k):
    positions = []
    for i in range(len(text)-len(pattern)+1):
        current_pattern = text[i:i+len(pattern)]
        if current_pattern == pattern or hammingDistance(current_pattern, pattern) <= k:
            positions.append(i)
    return len(positions)
```

```python
nu_list = ['A','C','T','G']

def immediateNeighbors(pattern):
    neighborhood = []
    for i in range(len(pattern)):
        neighborhood += [pattern[:i] + x + pattern[i+1:] for x in nu_list if x != pattern[i]]
    return set(neighborhood)    
```

```python
def neighbors(pattern, d):
    if d == 0: return set(pattern)
    if len(pattern) == 1:
        return {'A','C','G','T'}
    neighborhood = set()
    suffixNeighbors = neighbors(pattern[1:], d)
    for i in suffixNeighbors:
        if hammingDistance(pattern[1:], i) < d:
            for x in nu_list:
                neighborhood.add(x + i)
        else:
            neighborhood.add(pattern[0]+i)
    return neighborhood
```

```python
def computingFrequenciesWithMismatches(text, k, d):
    frequencyarray = []
    for i in range(4**k):
        frequencyarray.append(0)
    for i in range(len(text)-k+1):
        pattern = text[i:i+k]
        neighborhood = neighbors(pattern, d)
        for x in neighborhood:
            num_x = patternToNumber(x)
            frequencyarray[num_x] += 1
    return frequencyarray
        
```

```python
nu_dict = {'A' : 0, 'C' : 1, 'G': 2, 'T': 3}

def patternToNumber(pattern):
    if not pattern: 
        return 0
    last_nu = pattern[-1]
    prefix = pattern[:-1]
    return 4*patternToNumber(prefix) + nu_dict[last_nu]
```

```python
nu_dict_inverse = {0 : 'A', 1 : 'C', 2 : 'G', 3 : 'T'}

def numberToPattern(num, k):
    if k == 1:
        return nu_dict_inverse[num]
    prefixIndex = num//4
    r = num%4
    last_nu = nu_dict_inverse[r]
    return numberToPattern(prefixIndex, k-1)+last_nu
```

```python
def frequentWordsWithMismatches(text, k, d):
    frequentPatterns = set()
    neighborhoods = []
    index = []
    count = []
    for i in range(len(text)-k+1):
        for x in neighbors(text[i:i+k], d):
            neighborhoods.append(x)
    for i in range(len(neighborhoods)):
        pattern = neighborhoods[i]
        index.append(patternToNumber(pattern))
        count.append(1)
    index.sort()
    for i in range(len(neighborhoods)-1):
        if index[i] == index[i+1]:
            count[i+1] = count[i] + 1
    maxCount = max(count)
    for i in range(len(neighborhoods)):
        if count[i] == maxCount:
            pattern = numberToPattern(index[i], k)
            frequentPatterns.add(pattern)
    return frequentPatterns
            
        
```

```python
def frequentWordsWithMismatchesAndComplements(text, k, d):
    neighborhoods = []
    count = []
    new_frequency_array = dict()
    frequentWords = []
    index = []
    for i in range(len(text)-k+1):
        for x in neighbors(text[i:i+k], d):
            neighborhoods.append(x)
    for i in range(len(neighborhoods)):
        pattern = neighborhoods[i]
        index.append(patternToNumber(pattern))
    kmer_frequency = Counter(index)
    for index, count in kmer_frequency.items():
        comp_index = patternToNumber(complement(numberToPattern(index, k)))
        if comp_index in kmer_frequency.keys():
            new_frequency_array[index] = count + kmer_frequency[comp_index]
        else:
            new_frequency_array[index] = count
    maxCount = max(new_frequency_array.values())
    for index, count in new_frequency_array.items():
        if count == maxCount:
            frequentWords.append(numberToPattern(index, k))
    return frequentWords
    
    
    


            
```

```python
def count_with_mismatches(text, pattern, d):
    n = neighbors(pattern, d)
    count = 0
    for i in range(len(text)-len(pattern)+1):
        if text[i:i+len(pattern)] in n:
            count += 1
    return count
```
